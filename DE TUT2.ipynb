```python
#TXT
import requests
txt_url = 'https://www.gutenberg.org/files/1342/1342-0.txt'
response = requests.get(txt_url)
print(type(response)) #the actual type of data
txt_data = response.text
print(type(txt_data)) #the convereted type as string done by the library
print("First 200 characters of the TXT Data")
print(txt_data[:200]) #print first 200 characters
```

```python
#JSON
import json

#opening a json file
f = open('/content/drive/MyDrive/Colab Notebooks/WEEK 2/sample_users_with_id.json')
#returns JSON object as a dictionary
data = json.load(f)
#Iterating through the json
given_names = [user["given_name"] for user in data if "given_name" in user]
#print the names
print(given_names)
#closing file
f.close()
```
