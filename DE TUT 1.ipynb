```python
# Arithmetic Operations
a = 20
b = 4

# Addition
print(f"{a} + {b} = {a + b}")

# Subtraction
print(f"{a} - {b} = {a - b}")

# Multiplication
print(f"{a} * {b} = {a * b}")

# Division
print(f"{a} / {b} = {a / b}")

# Floor Division
print(f"{a} // {b} = {a // b}")

# Modulo (Remainder)
print(f"{a} % {b} = {a % b}")

# Exponentiation
print(f"{a} ** {b} = {a ** b}")
```

```python
# Comparison Operators
a = 10
b = 5

# Equal to
print(f"{a} == {b} is {a == b}")

# Not equal to
print(f"{a} != {b} is {a != b}")

# Greater than
print(f"{a} > {b} is {a > b}")

# Less than
print(f"{a} < {b} is {a < b}")

# Greater than or equal to
print(f"{a} >= {b} is {a >= b}")

# Less than or equal to
print(f"{a} <= {b} is {a <= b}")
```

```python
# Logical Operators
x = True
y = False

# AND operator
print(f"{x} and {y} is {x and y}")

# OR operator
print(f"{x} or {y} is {x or y}")

# NOT operator
print(f"not {x} is {not x}")
```

```python
text = "Data Engineering"
print (text.lower())
print(text[0:6]) #only the first 5 characters. Omits the 6th
print(text[::-1]) #reverse string through slicing
print(text[8:3:-1])#reverse slicing from 8th element to the 3rd(3 gets omitted so technically 8:2)
print(text.upper()) #printing a text in UPPER case
```
