```python
#total number of missing values per column.
sum_missing_pop=pop.isnull().sum()
sum_missing_pop.head(50)
```

```python
#investingating the % of missing values per column.
#sum where pop is null divided by length of pop & multiplied by 100(for %)
missing_pop=pop.isnull().sum()/len(pop)*100
missing_pop.head(50)
```

```python
#descriptive statistics for categorical variables. Here 'O' means object
pop.describe(include='O').T
```

```python
#creating a line plot for United Kingdom population data over the years
country_name = 'United Kingdom'

#filter data for said country
country_data = pop[pop['Country Name']==country_name]
```

```python
#select relevant columns
years = country_data.columns[4:] #years starts from the 4th column
pop_values = country_data[years].values[0]
```

```python
#create line plot
plt.figure(figsize=(12,6))
plt.plot(years,pop_values)
plt.title(f'Population Trend for{country_name}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#detecting outliers for 2016
pop_2016 = pop['2016']

#box plot
plt.figure(figsize=(10,6))
sns.boxplot(x=pop_2016)
plt.title('Box Plot of Population in 2016')
plt.xlabel('Population in 2016')
plt.show()
```

```python
#identifying outliers using IQR
Q1 = pop_2016.quantile(0.25)
Q3 = pop_2016.quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

outliers=pop_2016[(pop_2016<lower_bound) | (pop_2016 > upper_bound)]

#print countries with outlier population in 2016
print("Countries with outlier population in 2016:")
for country in pop[pop_2016.isin(outliers)] ['Country Name']:
  print(country)
```

```python
#scatter plot to identify outliers
plt.figure(figsize=(10,6))
plt.scatter(pop['Country Name'],pop_2016)
plt.title('Scatter Plot of Population in 2016')
plt.xlabel('Country Name')
plt.ylabel('Population in 2016')
plt.xticks(rotation=90)
plt.show()
```

```python
#creat a list of non country names
non_country=['Arab World', 'East Asia & Pacific (excluding high income)', 'Early-demographic dividend', 'East Asia & Pacific', 'Europe & Central Asia (excluding high income)', 'Europe & Central Asia', 'Euro area', 'European Union', 'Fragile and conflict affected situations', 'High income', 'Heavily indebted poor countries (HIPC)', 'IBRD only', 'IDA & IBRD total', 'IDA total', 'IDA blend', 'Indonesia', 'IDA only', 'Latin America & Caribbean (excluding high income)', 'Latin America & Caribbean', 'Least developed countries: UN classification', 'Low income', 'Lower middle income', 'Low & middle income', 'Late-demographic dividend', 'Middle East & North Africa', 'Middle income', 'Middle East & North Africa (excluding high income)', 'North America', 'OECD members',  'Pre-demographic dividend', 'Post-demographic dividend', 'South Asia', 'Sub-Saharan Africa (excluding high income)', 'Sub-Saharan Africa', 'East Asia & Pacific (IDA & IBRD countries)', 'Europe & Central Asia (IDA & IBRD countries)', 'Latin America & the Caribbean (IDA & IBRD countries)', 'Middle East & North Africa (IDA & IBRD countries)', 'South Asia (IDA & IBRD)', 'Sub-Saharan Africa (IDA & IBRD countries)', 'Upper middle income', 'World']

#remove rows where 'country name' is in the non_country
pop_filtered = pop[~pop['Country Name'].isin(non_country)]

#extract population data for the year 2016 from the filtered data
pop_2016_filtered = pop_filtered['2016']
```

```python
#visualizing filtered data outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=pop_2016_filtered)
plt.title('Box Plot of Population in 2016 (After Removing Non-Country Names)')
plt.xlabel('Population in 2016')
plt.show()
```

```python
# Identify outliers using the IQR method (after removing non-country names)
Q1_filtered = pop_2016_filtered.quantile(0.25)
Q3_filtered = pop_2016_filtered.quantile(0.75)
IQR_filtered = Q3_filtered - Q1_filtered
lower_bound_filtered = Q1_filtered - 1.5 * IQR_filtered
upper_bound_filtered = Q3_filtered + 1.5 * IQR_filtered

outliers_filtered = pop_2016_filtered[
    (pop_2016_filtered < lower_bound_filtered) | (pop_2016_filtered > upper_bound_filtered)
]

# Print the countries with outlier population in 2016 (after removing non-country names)
print("Countries with outlier population in 2016 (After Removing Non-Country Names):")
for country in pop_filtered[pop_2016_filtered.isin(outliers_filtered)]['Country Name']:
    print(country)
```

```python
# Create a scatter plot to visualize the relationship between population and outliers in 2016 (after removing non-country names)
plt.figure(figsize=(12, 6))
plt.scatter(pop_filtered['Country Name'], pop_2016_filtered)
plt.title('Scatter Plot of Population in 2016 (After Removing Non-Country Names)')
plt.xlabel('Country Name')
plt.ylabel('Population in 2016')
plt.xticks(rotation=90)
plt.show()
```
